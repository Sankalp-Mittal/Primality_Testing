```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gcd(a,b):
    while b:
        a,b = b,a%b
    return a
```

```python
def isPrime(n):
    if n == 1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('percentages.csv')

# Convert the DataFrame to a dictionary
percentages = df.set_index('n')['percentage'].to_dict()
```

```python
bins = np.arange(0,101,10)
hist, bin_edges = np.histogram(list(percentages.values()), bins=bins)

# Plot histogram
plt.bar(bin_edges[:-1], hist, width=8,align='edge', edgecolor='black')

# Convert y-axis to log scale
plt.yscale("log",base=10)

# Labels and title
plt.xlabel("Value Ranges")
plt.ylabel("Number of Values (Log Scale)")
plt.xticks(bins)  # Set tick marks at bin edges
plt.title("Distribution of Dictionary Values in Ranges (Log Scale)")

plt.show()
```
